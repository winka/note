## zip

### zip(*iterables, strict=False)

```python
friends= ["芸芸", "萱萱", "琳琳", "娜娜"]
stars = ["天秤", "金牛", "雙魚","雙子"]
likes = ["餅乾", "巧克力", "草莓", "蛋糕"]
```

```python
zipped = zip(friends, stars, likes)
list(zipped)
```

```python
datas = [('芸芸', '天秤', '餅乾'), 
         ('萱萱', '金牛', '巧克力'),
         ('琳琳', '雙魚', '草莓'),
         ('娜娜', '雙子', '蛋糕')]
zipped = zip(*datas)
print(list(zipped))
```

```python
answer = ['A', 'B', 'B', 'E', 'D', 'C']
wenwen = ['B', 'B', 'B', 'E', 'A', 'C']
```

```python
# 做對幾題
def check(answer,wenwen):
    return len( [1  for i in range(len(answer))  if answer[i] == wenwen[i] ] )
check(answer,wenwen)
```

```python
# 看第幾題做錯
def check(answer,wenwen):
    return [i+1  for i in range(len(answer))  if answer[i] == wenwen[i] ] 
check(answer,wenwen)
```

```python
def check(answer,wenwen):
    return len([1 for i , j in zip(answer,wenwen) if i == j] )
check(answer,wenwen)
```

## zip 常與 enumerate 連用

```python
def check(answer,wenwen):
    return [index+1 for  index, (i,j) in enumerate(zip(answer,wenwen)) if i == j] 
check(answer,wenwen)
```

```python
def rotateRight(arr):
    A = arr[::-1] #利用切片上下翻轉
    return list(map(list,(zip(*A)))) #行列互換，再利用map函數將zip內的元組轉列表
rotateRight(test)
```

## np.where

### Return elements chosen from x or y depending on condition.

```python
import numpy as np 
```

### If all the arrays are 1-D, where is equivalent to:

```python
arrays = [xv if c else yv
          for c, xv, yv in zip(condition, x, y)]
```

```python
a = np.arange(10)
a
```

```python
np.where(a < 5, a, 10*a)
# np.where(條件, 不符合條件要做甚麼(陣列), 不符合條件要做甚麼(陣列))
```

```python
np.where([[True, False], [True, True]],
         [[1, 2], [3, 4]],
         [[9, 8], [7, 6]])
```

## np.ogrid()

```python
x, y = np.ogrid[:3, :4]  
# 1.產生兩個二維陣列(廣播)  指派給 x, y
# 2.x, y 廣播變成相同大小
# 3.比較 x, y 得出 圈 1
# 4. if True 使用 圈 2 值
# 5. if FLASE 使用 圈 1 值
# 6. 合併

np.where(x < y, x, 10 + y) # np.where(條件, if T, if F)
```

```python
a = np.ogrid[0:5, 0:5] 
# list
# 產生一個串列內有兩個元素
# 每個元素都是 ndarray
# 第一個元素為 2維 5 * 1 陣列
# 第二個元素為 2維 1 * 5 陣列
```
