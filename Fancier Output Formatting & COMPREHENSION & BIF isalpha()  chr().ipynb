##  str.format() 

```python
yes_votes = 42_572_654 
no_votes = 43_132_495
percentage = yes_votes / (yes_votes + no_votes)


# 第一個變數傳給第一個大括弧 
# 第二個變數傳給第二個大括弧 
# 變數:格式 
# 按照指定的格式顯示
'{:-9} YES votes  {:2.2%}'.format(yes_votes, percentage)
#　2.2% 意思在下面
```

## 生成式 COMPREHENSION
#### https://mathspp.com/blog/pydonts/list-comprehensions-101

```python
# 前面只能有一個運算式
[n**2 for n in range(10)]
```

```python
[n ** 2 for n in range(10) if n %2 == 0]
```

```python
words = 'this is sparta!'.split()
[word.upper() for word in words]
```

```python
[word.upper() for word in words if word == word.lower()]
```

```python
words2 = 'to be or not to be, that is the question'.split() 
words2
```

```python
[len(word) for word in words2]
```

```python
[len(word) for word in words2 if word.isalpha()] # 檢測字母是否由字母組成
```

##  str.isalpha()

### Python isalpha() 方法检测字符串是否只由字母组成。

```python
str1 = "runoob";
print(str1.isalpha());

str2 = "runoob菜鸟教程";
print(str2.isalpha());

str3 = "this is string example....wow!!!";
print(str3.isalpha());

# True
# True
# False
```

```python
from random import randint
r, g, b = [randint(0, 255) for _ in range(3)]
f'{r,g,b}'
```

```python
from string import ascii_lowercase, ascii_uppercase
from random import choice
# join 加入 
# 前面通常是''(str)
# choice() 從括號裡面選擇 
# for 做16次
"".join([choice(ascii_lowercase + ascii_uppercase) for _ in range(16)])
```
