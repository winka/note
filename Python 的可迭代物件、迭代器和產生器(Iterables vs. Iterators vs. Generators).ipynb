## Iterables

### 可迭代物件的定義是：
### 1. 用來產生迭代器的物件，回傳其包含的所有元素，但它不必是資料結構。
### 2. 可迭代物件通常會實作 __iter__() 和 __next__() 這兩個方法，使得該物件同時是可迭代及迭代器。

```python
### __iter__() 讓可跌代物件(iterable)產生跌代器(iterator)
### __next__() 讓跌代器(iterator)產生下一個元素
```

### x 是「可迭代」的物件，而 y 和 z 則是兩個互不相關的迭代器實例，
### 它們從「可迭代物件」x 來產出資料。y 和 z 會有指向 x 某個元素的「狀態」，並依照某些規則輸出 x 裡的元素。

### X 透過類似 iter() 的機制變成迭代器，再經過類似 next() 的機制取得 X 內部的所有元素。

## 產生器(Generator)

### 產生器是一種特別的迭代器。

### 產生器讓你不用實作 __iter__() 和 __next__() 也能建立迭代器。

### 產生器都是迭代器(iterator)，但反過來不一定成立
### 承上點，產生器能懶洋洋地輸出值 (lazy factory)

## 產生器的種類
### 有兩種產生器：
### 1.產生器函數 (generator function)
### 2.產生器表示式 (generator expression)
